```python
import pandas as pd
import numpy as np
data = pd.read_csv('dailymintemps.csv',header=0,index_col=0)
data.head()
```

```python
data.values
```

```python
temps = pd.DataFrame(data.values)
temps1 = temps.shift(1)
print(temps)
print(temps1)
```

```python
df = pd.concat([temps1,temps],axis=1)
df.columns = ['t-1','t+1']
df
```

```python
df = pd.concat([temps.shift(3),temps.shift(2),temps.shift(1),temps],axis=1)
df.columns = ['t-3','t-2','t-1','t-1']
df
```

```python
window = temps.expanding()
dataframe = pd.concat([window.min(), window.mean(), window.max(), temps.shift(-1)], axis=1)
dataframe.columns = ['min', 'mean', 'max', 't+1']
print(dataframe.head(5))
```
